```python
from typing import List
from matplotlib import pyplot
```

```python
recalls: List[float] = []
precisions: List[float] = []
with open("eval/slow3.csv", "r") as f:
    for line in f.readlines():
        line_values = line.split(',')
        if line_values[0].startswith("conf"):
            continue
        recalls.append(float(line_values[12]))
        precisions.append(float(line_values[7]))
```

```python
pyplot.plot(recalls, precisions, marker='.', label='Logistic')
# axis labels
pyplot.xlabel('Recall')
pyplot.ylabel('Precision')
# show the legend
pyplot.legend()
# show the plot
pyplot.show()
```
